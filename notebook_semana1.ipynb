```python
 # ENTRADA: ENTRA ANCHO Y ALTO DE LA MADERA EN METROS
 ancho = input("ingresa el ancho de la ventana en metros")
 ancho = float(ancho)
 alto = input("ingresa el alto de la ventana en metros")
 alto = float(alto)

# PROCESO: calcula longitud de la madera (lm) en pies y el area del vidrio en metros cuadrados
lm = 2 * (ancho + alto) * 3.25
av = 2 * (ancho * alto)
# SALIDA: sale la longitud de la madera en pies el area del vidrio en metros cuadrados
print("longitud de la madera es ", lm," pies de longitud")
print("el area de vidrio es " ,av , " metros cuadrados")

```
