## ¿Qué es Python?

Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser "legible", lo que permite a profesionales que no son programadores de oficio aprenderlo rápido para resolver problemas específicos.

## ¿Por qué es importante en Ingeniería Agroindustrial?

En la agroindustria, pasamos de manejar materias primas en el campo a procesos de transformación complejos en la planta. Python brilla aquí por tres razones:

### 1. Cálculo del Índice de Rendimiento en Extracción

Este código ayuda a calcular qué porcentaje de producto terminado obtuviste a partir de una materia prima.

```
# Cálculo de rendimiento de aceite esencial
materia_prima = 500  # kg de naranja
producto_final = 2.5  # kg de aceite extraído

rendimiento = (producto_final / materia_prima) * 100

print(f"El rendimiento de la extracción es del {rendimiento}%")
```

### 2. Conversión de Unidades de Temperatura (Pasteurización)

Muy útil cuando trabajas con equipos que vienen en diferentes sistemas de medidas.

```
def celsius_a_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

temp_proceso = 72  # Grados Celsius para pasteurización rápida
print(f"La temperatura de proceso en Fahrenheit es: {celsius_a_fahrenheit(temp_proceso)}°F")
```

### 3. Clasificación Automática por Peso

Imagina un sensor en una banda transportadora que clasifica frutas según su peso para empaque.

```
peso_fruta = 155  # gramos

if peso_fruta < 100:
    categoria = "Pequeña (Descarte o Jugo)"
elif 100 <= peso_fruta <= 200:
    categoria = "Mediana (Exportación)"
else:
    categoria = "Grande (Mercado Local)"

print(f"La fruta de {peso_fruta}g clasifica como: {categoria}")
```
